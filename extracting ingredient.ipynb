```python
import pandas as pd
from collections import Counter
```

```python
data = pd.read_excel(r'C:\Users\7fno\Desktop\pizza_sales_excel_file.xlsx')
```

```python
data
```

```python
data['pizza_ingredients'].value_counts()
```

```python
ingredient = [x.strip() for y in data['pizza_ingredients'].str.split(',') for x in y]
```

```python
ingredient
```

```python
count_ingredient = Counter(ingredient)
```

```python
df=  pd.DataFrame(count_ingredient.items(),columns=['ingredient','count'])
```

```python
df
```

```python
df = df.sort_values(['count'],ascending=False).reset_index(drop=True)
df
```

```python
df.to_csv(r'ingredient.csv')
```
